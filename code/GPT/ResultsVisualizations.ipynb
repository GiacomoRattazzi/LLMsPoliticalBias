```python
import pandas as plt
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Function to plot stance distribution
def plot_stance_distribution(data):
    # Count of unique values in categorical columns
    unique_values_count = data.select_dtypes(include=['object']).nunique()

    # Reordering the 'Predicted Stance' categories
    data['Predicted Stance'] = pd.Categorical(data['Predicted Stance'], categories=['left', 'center', 'right'], ordered=True)

    # Count of Predicted Stances
    plt.figure(figsize=(10, 6))
    ax = sns.countplot(x='Predicted Stance', data=data, palette="coolwarm", order=['left', 'center', 'right'])
    plt.title('Count of Predicted Stances')
    plt.xlabel('Predicted Stance')
    plt.ylabel('Count')

    # Adding value labels on top of each bar
    for p in ax.patches:
        ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha = 'center', va = 'center',
                    xytext = (0, 10),
                    textcoords = 'offset points')
    plt.show()

# Function to plot BERT Score distribution
def plot_bert_score_distribution(data):
    plt.figure(figsize=(10, 6))
    sns.histplot(data['BERT Score'], kde=True, color='#87D0BE')
    plt.title('Distribution of BERT Scores')
    plt.xlabel('BERT Score')
    plt.ylabel('Frequency')
    plt.xlim(0.1, 0.9)
    plt.show()
```

```python
# Loading datasets
file_path_base= 'datagptBase.csv'
file_path_left= 'datagptLeft.csv'
file_path_right= 'datagptRight.csv'
data_base = pd.read_csv(file_path_base)
data_left = pd.read_csv(file_path_left)
data_right = pd.read_csv(file_path_right)
```

**Base Model Results**

```python
plot_stance_distribution(data_base)
```

```python
plot_bert_score_distribution(data_base)
```
